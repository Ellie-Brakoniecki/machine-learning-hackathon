## Data Preprocessing

```python
# Choose features

Features = ['Year Built', 'Central Air','Roof Style', 'Exterior 1st', 'Exterior 2nd',
            'Heating','Electrical','Kitchen Qual','Garage Type','Garage Finish' , 'Bsmt Qual', 'Overall Qual']

df_clean = AMESdf[Features]
```

```python
# Label Modern vs historical

df_clean['Modern'] = df_clean['Year Built'].apply(lambda x: 1 if x > 2000 else 0)
df_clean.head()
```

```python
# Drop Missing Values

df_clean = df_clean.dropna()
df_clean.shape
```

```python
# encoding 

# Categorical columns

catagorical_cols = [col for col in df_clean if df_clean[col].dtype =='object']
print(catagorical_cols)
#apply encoding 

df_clean = pd.get_dummies(df_clean,columns=catagorical_cols)

df_clean.head()
```

```python
# remove year built

df_clean = df_clean.drop('Year Built', axis = 1)
```

```python
# Split Dataset
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df_clean, test_size=0.2, random_state= 42,stratify=df_clean['Modern'])
```

## Model Building

```python
# Random forest

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# data for training

X_train = train_df.drop('Modern', axis=1)
y_train = train_df['Modern']

# testing data

X_test = test_df.drop('Modern', axis=1)
y_test = test_df['Modern'] 

# Instantiatio mmodel

rf_model = RandomForestClassifier(random_state=9)

# Train model

rf_model.fit(X_train,y_train)

#Predict

rf_predictions = rf_model.predict(X_test)

```

```python
# Gradient boost 

from sklearn.ensemble import GradientBoostingClassifier

# Instantiate model 

gb_model = GradientBoostingClassifier()

# Train MOdel 

gb_model.fit(X_train,y_train)

# predictions

gb_predictions = gb_model.predict(X_test)
```

## Model Evaluation

```python
# Random forest 

#Metrics
print(accuracy_score(y_test, rf_predictions))
print(classification_report(y_test, rf_predictions))
```

```python
# Random forest 

#Metrics
print(accuracy_score(y_test, gb_predictions))
print(classification_report(y_test, gb_predictions))
```
